```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sns.set()
np.random.seed(0)
```

```python
uniform_data = np.random.rand(3,3)
print(uniform_data)
#热度图绘制
heatmap = sns.heatmap(uniform_data)
```

```python
# 设置热度图颜色取值范围
ax =sns.heatmap(uniform_data,vmin = 0.2,vmax = 0.5)
```

```python
normal_data = np.random.randn(3,3)
print(normal_data)
sns.heatmap(normal_data,center = 0)
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
# print(help(flights.pivot))
#转置矩阵，将month作index，year作列，passengers作为表格数值的矩阵
flights = flights.pivot("month","year","passengers")
print(flights)
ax = sns.heatmap(flights)
```

```python
#,fmt = "d"可以直接显示数值，而不用科学计数法
ax = sns.heatmap(flights,annot = True,fmt = "d")
```

```python
#间隔线宽度
ax = sns.heatmap(flights,linewidth =0.5)
```

```python
#指定调色板YlGnBu yellow->green->blue
ax = sns.heatmap(flights,cmap = "YlGnBu")
```

```python
# 隐藏指示条
ax = sns.heatmap(flights,cbar =False)
```
