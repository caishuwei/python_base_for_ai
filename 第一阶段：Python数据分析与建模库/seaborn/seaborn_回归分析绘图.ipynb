```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats,integrate
%matplotlib inline
```

```python
sns.set(color_codes = True)
```

```python
#设置好随机数种子
np.random.seed(sum(map(ord,"regression")))
```

```python
#使用seaborn提供的tips数据集 饭店顾客消费
tips = sns.load_dataset("tips")

columns = tips.columns
col_names = np.array([columns,columns])
col_names[1,0] = "客户消费"
col_names[1,1] = "小费"
col_names[1,2] = "性别"
col_names[1,3] = "是否抽烟"
col_names[1,4] = "星期"
col_names[1,5] = "消费时段"
col_names[1,6] = "人数"
print(col_names.T)

print(tips.head())
```

### regplot()和lmplot()都可以绘制回归关系，推荐regplot()

```python
#lmplot功能更多，规范也更多，限制比较多
#使用regplot绘制拟合出的回归线，消费金额更大的情况下，给的小费更多
sns.regplot(x = "total_bill",y = "tip",data = tips)
```

```python
sns.regplot(data = tips,x = "size",y = "tip")
```

```python
#对于x轴类型较少的情况，可以加上x_jitter对x轴的值进行随机偏移
sns.regplot(data = tips,x = "size",y = "tip",x_jitter = 0.05)
```
