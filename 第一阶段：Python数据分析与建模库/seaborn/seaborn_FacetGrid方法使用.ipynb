```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats,integrate
from pandas import Categorical

%matplotlib inline
```

### 当数据集中有很多子集需要展示，就可以使用FacetGrid

```python
sns.set(style = "white",color_codes=True)
```

```python
np.random.seed(sum(map(ord,"axis_grids")))
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
#先占位
g = sns.FacetGrid(tips,col = "time")
```

```python
g = sns.FacetGrid(tips,col = "time")
# map方法第一个参数指定绘图方法，这里用了plt库的hist方法，也可以用sns的方法
g.map(plt.hist,"tip")
```

```python
g = sns.FacetGrid(tips,col = "sex",hue = "smoker")
#alpha指定每个点绘制时的透明度
g.map(plt.scatter,"total_bill","tip",alpha = 0.7)
g.add_legend()
```

```python
g = sns.FacetGrid(tips,col = "time",row = "smoker",margin_titles = True)
#fit_reg 是否绘制回归线，x_jitter 绘制时左右偏移以便观察数据分布，color 指定颜色
g.map(sns.regplot,"size","total_bill",color = "0.3",fit_reg=True,x_jitter = 0.1)
```

```python
g = sns.FacetGrid(tips,col = "day",size = 4,aspect = 0.5)
# 
# 出现如下警告，大意是使用直方图进行绘制时，x坐标的排序未明，这里应该是根据sex列数据中Female 和Male的出现顺序进行了排序
# C:\Users\caisw\Anaconda3\lib\site-packages\seaborn\axisgrid.py:703: UserWarning: Using the barplot function without specifying `order` is likely to produce an incorrect plot.
#   warnings.warn(warning)

#添加排序属性，指定sex枚举值排序（如果sex是整型123这些就没有排序问题，字符串最好还是指定一下枚举顺序）
g.map(sns.barplot,"sex","total_bill",order = ["Female","Male"])
```

```python
#图按日期分为4个图，画图前根据每天的消费次数进行排序，这个排序后面将应用于FacetGrid，指定图的顺序
ordered_days = tips.day.value_counts().index
print(ordered_days)

#自己指定排序
ordered_days = Categorical(["Thur","Fri","Sat","Sun"])
#直接字符串列表也是可以的。。。
ordered_days = ["Thur","Fri","Sat","Sun"]

g = sns.FacetGrid(tips,row = "day",row_order = ordered_days,size = 1.7,aspect = 4)
#y轴没有分类，那就不需要排序了，然而还是会有警告，添加一个空排序
g.map(sns.boxplot,"total_bill",order = [])
#有点奇葩，y轴上的排序，居然越靠前越远离原点，，，，
# g.map(sns.boxplot,"total_bill","sex",order = ["Male","Female"])
```
