```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats,integrate
%matplotlib inline
```

```python
sns.set(style = "whitegrid",color_codes=True)
```

```python
np.random.seed(sum(map(ord,"categorical")))
```

```python
titanic = sns.load_dataset("titanic")
#窝草，seaborn里面有泰坦尼克号数据的吗
print(titanic.head())
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

```python
#消费量与星期的关系
sns.stripplot(x = "day",y = "total_bill",data = tips)
```

#### 重叠是很常见的现象，但重叠影响我观察数据的量了

```python
#jitter将数据进行随机偏移，方便观察
sns.stripplot(x = "day",y = "total_bill",data = tips,jitter = True)
```

```python
#更夸张的一个图，左右偏移保证点不重合
sns.swarmplot(x = "day",y = "total_bill",data = tips)
```

```python
#hue可以根据sex的类型通过颜色将点的性别表示出来
sns.swarmplot(x = "day",y = "total_bill",data = tips,hue = "sex")
```

```python
sns.swarmplot(x = "total_bill",y = "day",data = tips,hue = "time")
```

#### 盒图

* IQR即统计学概念四分位距，第1/4分位与第3/4分位之间的距离N = 1.5IQR 
* 如果一个值>Q3+N或 < Q1-N,则为离群点

```python
sns.boxplot(x = "day",y = "total_bill",hue = "time",data = tips)
```

```python
#小提琴图，与swarmplot类似，数据越集中的地方越宽，通常都是一个小提琴的模样，所以叫小提琴图
sns.violinplot(x = "total_bill",y = "day",hue = "time",data = tips)
```

```python
#split将hue分开绘制的小提琴图合并在一起，切分左右进行表示
sns.violinplot(x = "day",y = "total_bill",hue = "sex",data = tips,split= True)
```

```python
#在同一个图表中绘图
sns.violinplot(x ="day",y = "total_bill",data = tips,inner = None)
sns.swarmplot(x = "day",y = "total_bill",data = tips,color = "w",alpha = 0.5)
```

#### 显示值的集中趋势可以用条形图

```python
sns.barplot(x = "sex",y = "survived",hue = "class",data = titanic)
```

#### 点图可以更好的描述变化差异

```python
#呈现男女之间的生存差异
sns.pointplot(x = "sex",y = "survived",hue = "class",data = titanic)
```

```python
#设置其他样式
#palette为hue类别设置不同颜色，类型dict
#markers为hue类别设置不同标记，^表示三角形，o表示圆形
#linestyles，设置线条样式，不多说了

sns.pointplot(x = "class",y = "survived",hue = "sex",data = titanic,
              palette = {"male":"g","female":"m"},
              markers = ["^","o"],
              linestyles = ["-","--"]
             )
```

```python
#横向盒图
sns.boxplot(data = iris,orient = "h")
```
