### 调色板
* 颜色很重要
* color_palette()能传入任何Matplotlib所支持的颜色
* color_palette()不写参数则默认颜色
* set_palette()设置所有图的颜色

### 分类色板

```python
#获取调色板默认颜色（6个），当需要画八条线的时候会循环选择颜色
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

6个默认的颜色循环主题：deep muted pastel bright dark colorblind

### 圆形画板

当你有六个以上的分类要区分时，最简单的方法就是在一个圆形的颜色空间中画出均匀间隔的颜色（这样的色调会保持亮度和饱和度不变）。这是大多数的当他们需要使用比当前默认颜色循环中设置的颜色更多时的默认方案。

最常用的方法是使用hls的颜色空间，这是RGB值的一个简单转换

```python
#当然有时为了在绘制八条线条时不出现颜色重复，需要使用8个不同的颜色进行绘制
#通过hls取得8个均匀间隔的颜色
sns.palplot(sns.color_palette("hls",8))
```

```python
sns.palplot(sns.color_palette("hls",12))
```

```python
data = np.random.normal(size =(20,8))+np.arange(8)/2
sns.boxplot(data = data,palette = sns.color_palette("hls",8))
```

## hls_palette()函数可以控制颜色的饱和度和亮度
* l 亮度lightness(默认值0.5 [0,1])
* s 饱和saturation(默认值0.5 [0,1])

```python
sns.palplot(sns.hls_palette(8,l = 0.5,s = 0.5))
sns.palplot(sns.hls_palette(8,l = 0.5,s = 1))
sns.palplot(sns.hls_palette(8,l = 0.1,s = 0.5))
```

```python
#当需要成对的颜色时，可以使用Paired，生成的颜色一深一浅
sns.palplot(sns.color_palette("Paired",8))
```

### xkcd颜色来命名颜色
xkcd包含了一套众包努力的针对随机RGB色的命名。产生了954个可以随时通过xkcd_rgb字典中调用的命名颜色

```python
#绘制线条，传入xarray ,yarray,指定颜色，指定线宽
#xkcd针对颜色命名，如果不是很熟悉他的命名的话，不会很好用、、、、
plt.plot([0,1],[0,1],sns.xkcd_rgb["pale red"],lw = 3)
plt.plot([0,1],[0,2],sns.xkcd_rgb["medium green"],lw = 3)
plt.plot([0,1],[0,3],sns.xkcd_rgb["denim blue"],lw = 3)
```

```python
colors = ["windows blue","amber","greyish","faded green","dusty purple"]
sns.palplot(sns.xkcd_palette(colors))
```

### 连续画板
色彩随数据变换，比如数据越来越总要则颜色越来越深

```python
#直接指定颜色，会由白色渐变到该颜色，并等分（默认6）选取每段的中心颜色
sns.palplot(sns.color_palette("Blues"))
```

```python
#使用xkcd颜色名称衔接_r，表示由指定颜色渐变到白色
sns.palplot(sns.color_palette("BuGn_r",8))
```

```python
#黑色渐变到目标颜色
sns.palplot(sns.color_palette("cubehelix",8))
```

### cubehelix_palette

色调线性变换

```python
sns.palplot(sns.cubehelix_palette(8,start = .5,rot = -.75))
```

```python
sns.palplot(sns.cubehelix_palette(8,start = .75,rot = -.150))
```

### light_palette()和dark_palette()调用定制连续调色板

```python
sns.palplot(sns.light_palette("green"))
```

```python
sns.palplot(sns.dark_palette("purple"))
```

```python
#反转渐变
sns.palplot(sns.light_palette("navy"))
sns.palplot(sns.light_palette("navy",reverse = True))
```

```python
#multivariate_normal，用于生成矩阵，
x,y = np.random.multivariate_normal([0,0],[[1,-0.5],[-0.5,1]],size = 300).T
pal = sns.dark_palette("green",as_cmap = True)
sns.kdeplot(x,y,cmap = pal)
```

```python
sns.palplot(sns.light_palette((210,90,60),input = "husl"))
```
