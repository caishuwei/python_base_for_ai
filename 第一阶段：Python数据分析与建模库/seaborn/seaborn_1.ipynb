```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
#写完代码自动显示图表不用plt.show()
%matplotlib inline
```

```python
#seaborn也是一个可视化库，是基于matplotlib的基础上进行封装的
#它提供了大量模板用于画图
```

```python
def sinplot(flip = 1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
#seaborn提供的 默认设置
sns.set()
sinplot()
```

5种主题风格
*  darkgrid
*  whitegrid
*  dark
*  white
*  ticks

```python
sns.set_style("whitegrid")
data = np.random.normal(size = (20,6))+np.arange(6)/2
sns.boxplot(data = data)
```

```python
sns.set_style("darkgrid")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("ticks")
#去掉上边与右边的刻度（这个框架现在默认就是去掉的。。。）
sns.despine()
sinplot()
```

```python
sns.violinplot(data)
#设置轴偏移量
sns.despine(offset = 10)
```

```python
sns.set_style("whitegrid")
#palette设置色调，deep为深色调
sns.boxplot(data = data,palette = "deep")
#隐藏左边的轴
sns.despine(left = True)
```

```python
#with域实现局部样式应用，类似于canvas的状态保存
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
    
plt.subplot(212)#居然还可以直接用10进制，每位数代表维度 等价于2,1,2
sinplot(-1)
```

```python
#将风格重置
sns.set()
```

指定context,内容样式会改变，style主要是影响表格
而context影响绘制内容

下面这些样式会影响绘制线条粗细
* paper 纸张
* talk 对话
* poster 海报样式
* notebook 笔记

font_scale 参数可以修改字体大小（缩放倍数）
rc={"lines.linewidth":2.5}修改线条宽度

```python
sns.set_context("paper")
plt.figure(figsize = (8,6))
sinplot()
```

```python
sns.set_context("talk")
plt.figure(figsize = (8,6))
sinplot()
```

```python
sns.set_context("poster")
plt.figure(figsize = (8,6))
sinplot()
```

```python
sns.set_context("notebook",font_scale = 2,rc = {"lines.linewidth":2.5})
# print(help(sns.set_context))
plt.figure(figsize = (8,6))
sinplot()
```
