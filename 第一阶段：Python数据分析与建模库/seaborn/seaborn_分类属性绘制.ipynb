### 多层面板分类图

### seaborn.factorplot
#### factorplot(x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean at 0x000001A2DB682C80>, ci=95, n_boot=1000, units=None, order=None, hue_order=None, row_order=None, col_order=None, kind='point', size=4, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)

### Parameters

* x,y,hue 数据集变量，变量名
* data 数据集 数据集名
* row,col 更多分类变量进行平铺显示 变量名
* col_wrap 每行的最高平铺数 整数
* estimator 在每个分类中进行矢量到变量的映射 矢量
* ci 置信区间 浮点数或None
* n_boot 计算置信区间时使用的引导迭代次数 整数
* units 采样单元的标识符，用于执行多级引导和重复测量设计 数据变量或向量数据
* order,hue_order 对应排序列表 字符串列表
* row_order,col_order 对应排序列表，字符串列表
* kind 可选：
 + point 点图（默认）
 + bar 柱形图
 + count 频次
 + box 盒图
 + violin 小提琴图
 + strip 散点图
 + swarm 散点图（左右偏移实现点不重合）
* size 每个面的高度（英寸）标量
* aspect 纵横比 标量
* orient 方向（v/h）
* color 颜色 matplotlib颜色
* palette 调色板 seaborn颜色色板或字典
* legend hue的信息面板，即图例 Ture/False
* legend_out 是否扩展图形，并将图例绘制在中心右边 True/False
* share{x,y} 共享轴线 True/False



```python
#默认绘制类型是点图
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips)
```

```python
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips,kind = "bar")
```

```python
#col 分多列进行分析
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",col = "time",data = tips,kind = "swarm")
```

```python
sns.factorplot(x = "time",y = "total_bill",hue = "smoker",col = "day",data = tips,kind = "box",size = 4,aspect = 0.5)
```
