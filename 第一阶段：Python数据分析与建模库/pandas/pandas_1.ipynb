```python
#打印e结尾的列值
col_names = titanic_data.columns.tolist()
print("-----所有列名------")
print(col_names)

select_cols = []
for c in col_names:
    if c.endswith("e"):
        select_cols.append(c)

print("-----选择的列名------")
print(select_cols)

print("------过滤不必要列后的DataFrame-------")
print(titanic_data[select_cols])
```
