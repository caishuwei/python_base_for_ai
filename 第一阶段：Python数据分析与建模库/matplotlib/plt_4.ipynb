```python
import pandas as pd
import numpy as np
import matplotlib as mpl

from matplotlib import pyplot as plt
```

```python
# 柱形图，与条形图外观类似，条形图表示某个值数量多少
# 而柱形图用于表示某个区间内的数量有多少

# print(help(np.random.random))
# 生成一组数据，数值范围0~10，取到小數點后1位，数量500
data = np.random.random((500,))*100//1/10
# print(data)

# print(help(plt))
fig = plt.figure(figsize=(15,10))
# print(help(fig))

#条形图
# 0.0, 0.1, 0.2 ,,,,9.9 一共要100条
value_types = np.linspace(0, 9.9, 100)
value_count = np.zeros(100)
index = 0
for i in data:
    index= int(i*10)
    value_count[index] = value_count[index]+1
    
print(value_count)

ax1 = fig.add_subplot(2,1,1)
# print(help(ax1))
ax1.bar(value_types,value_count,0.02)
ax1.set_xlabel("0.0~9.9")
ax1.set_ylabel("count")
ax1.set_title("0.0~9.9内随机500个数，条形图显示分部分布",fontproperties = "SimHei")

ax2 = fig.add_subplot(2,1,2)
#柱形图，将100个数（0~10）分成50块，第一块表示0~0.2区间的数量
#bins表示当前显示的柱的数量，默认值10
#range指定要显示的x轴区间
ax2.hist(data,range = (4,8),bins =40)

plt.show()
```

```python
#盒图,用于查看数据分布情况
fig,ax = plt.subplots(figsize = (10,10))
ax.boxplot(data)
ax.set_xticklabels(["500个随机数的分布情况"],fontproperties = "SimHei")
plt.show()
#由于是随机取值，取了500个数，所以分布比较平均了
#仔细看可以看出，125~250个最小的数跨越范围是比较大的，说明落在这个区间里的数密度较小
#盒图也叫4分图，将数据进行排序，平分成4块，可以比较1/4数量的数所占区间的大小，所占空间小的盒子，说明密度高，
#如果是学生考试成绩，四分图显示出来的前后两个盒跨度大，中间两个盒跨度小，大约就在60~80中间
```
