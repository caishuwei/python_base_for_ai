```python
#-*- coding = utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#条形图
```

```python
#这张表是我根据视频里面打印的数据补充的一行数据。。。
reviews = pd.read_csv("csv/film.csv")
#要打印的列
cols = ["FILM","RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]
norm_reviews = reviews[cols]
print(norm_reviews[:1])
```

```python
#绘制条形图
#读取需要的列值
num_cols = ["RT_user_norm","Metacritic_user_nom","IMDB_norm","Fandango_Ratingvalue","Fandango_Stars"]
#视频里面的ix方法过时了，推荐用loc
# print(help(norm_reviews.loc))
bar_hegihts = norm_reviews.loc[0,num_cols].values
print(bar_hegihts)
#设置柱形图位置
bar_positions = np.arange(len(num_cols))+1
print(bar_positions)
#取得绘图对象
#subplots函数返回一个元组，包含Figure与AxesSubplot
#Figure前面讲过用于添加子图，AxesSubplot用于绘制柱形图
fig,ax= plt.subplots()
print(type(fig))
print(type(ax))
#通过bar方法设置,align属性指定柱形图绘制起点，edge是左边开始，center是柱形图中心点在设置的position上
#什么鬼框架，看视频里面align默认值应该是edge....,现在是center,来搞笑的吗
# print(help(ax.bar))
#柱形图居中绘制，宽度0.5
ax.bar(bar_positions,bar_hegihts,0.5,align = "center")
ax.set_xticks(bar_positions)
ax.set_xticklabels(num_cols,rotation = 45)

ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Avengers:Age of Ultron (2015)")

plt.show()
```

```python
#横向柱形图
fig,ax= plt.subplots()
#直接通过barh方法就可以横向绘制
ax.barh(bar_positions,bar_hegihts,0.5,align = "center")
ax.set_yticks(bar_positions)
ax.set_yticklabels(num_cols)

ax.set_ylabel("Rating Source")
ax.set_xlabel("Average Rating")
ax.set_title("Average User Rating For Avengers:Age of Ultron (2015)")

plt.show()
```

```python
#散点图
# AxesSubplot.scatter

#豆瓣猫眼采集的一些数据
#直接读取设置编码为utf-8也会报错

douban_maoyan = pd.read_csv("csv/douban_maoyan.csv")
print(douban_maoyan)

fig ,ax = plt.subplots()
# print(help(fig))
# fig用于设置图表属性，这里设置宽高多少英寸。。。md居然是这个单位
fig.set_size_inches(10,10)
# print(help(ax.scatter))
# x轴数据，y轴数据
ax.scatter(douban_maoyan["豆瓣"],douban_maoyan["猫眼电影"])

# print(help(ax.set_xlabel))
# import matplotlib
# print(help(matplotlib.text.Text))
# print(help(matplotlib.artist.Artist))
# print(help(matplotlib.font_manager.FontProperties))

# fontproperties 是字体属性配置，使用SimHei可以显示中文
ax.set_xlabel("豆瓣",fontproperties='SimHei')
ax.set_ylabel("猫眼电影",fontproperties='SimHei')

#从文档里面找到的方法，设置xy轴的数据范围
#也可以使用set_xlim 这个与bound的区别是，bound是边界，如果要绘制的点超出当前边界，则会更新边界使绘制的东西能显示
#set_xlim则是限制图表大小，超出区域的数据绘制了也看不到
ax.set_xbound(0,10)
ax.set_ybound(0,10)
plt.show()

#从散点图分布情况上看，猫眼用户对电影的评分都是比较高的
```

```python
#也可以使用子图同时显示多个散点图
# print(help(plt.subplots))
# print(help(plt.figure))
fig =  plt.figure() 
fig.set_size_inches(10,10)

film_pos = np.arange(8)+1
ax0 = fig.add_subplot(2,2,2)
# print(help(ax0))
ax0.scatter(douban_maoyan["豆瓣"],film_pos,color = "#FF0000",label = "豆瓣")
ax0.scatter(douban_maoyan["猫眼电影"],film_pos,color = "#0000FF",label = "猫眼电影")
ax0.set_xbound(0,10)
ax0.set_ylabel("电影",fontproperties='SimHei')
ax0.set_xlabel("评分（0~10）",fontproperties='SimHei')
ax0.set_yticks(film_pos)
ax0.set_yticklabels(douban_maoyan["电影"],fontproperties='SimHei')
legend = ax0.legend(loc = "best")

#遍历所有图例文本，设置字体属性，否则中文无法显示。。。。艹不是这么麻烦吧
for text in legend.texts:
    text.set_font_properties("SimHei")
# print(help(legend.texts[0]))

# print(help(fig))
ax1 = fig.add_subplot(2,2,3)
# print(help(ax1))
ax1.scatter(douban_maoyan["豆瓣"],douban_maoyan["猫眼电影"])
ax1.set_xbound(0,10)
ax1.set_ybound(0,10)
ax1.set_xlabel("豆瓣",fontproperties='SimHei')
ax1.set_ylabel("猫眼电影",fontproperties='SimHei')

ax2 = fig.add_subplot(2,2,4)
ax2.scatter(douban_maoyan["猫眼电影"],douban_maoyan["豆瓣"])
ax2.set_xbound(0,10)
ax2.set_ybound(0,10)
ax2.set_ylabel("豆瓣",fontproperties='SimHei')
ax2.set_xlabel("猫眼电影",fontproperties='SimHei')


plt.show()
```
