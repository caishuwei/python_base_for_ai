```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#添加子图
fig = plt.figure()
ax1 = fig.add_subplot(4,3,1)#当前子图在4行3列的模式下位于第一个元素位置
ax2 = fig.add_subplot(4,3,2)
ax3 = fig.add_subplot(4,3,6)
plt.show()
```

```python
fig = plt.figure(figsize = (5,5))#设置折线图大小
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(np.random.randint(1,5,5),np.arange(5))
ax2.plot(np.arange(10),np.arange(10))
plt.show()
```

```python
unrate = pd.read_csv("csv/unrate.csv")
#利用pandas to_datetime函数将日期字符串转为日期对象，替换DATE列
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
#添加新的列MONTH,从之前转换的日期中读取month
unrate["MONTH"] = unrate["DATE"].dt.month
fig = plt.figure(figsize = (6,3))

#绘制1948年的未就业率折线与1949年未就业率折线图
plt.plot(unrate[0:12]["MONTH"],unrate[0:12]["VALUE"],c="red",label = "1948")
#参数c (color) 可以使用FF0000 ,label指定线条标签
plt.plot(unrate[12:24]["MONTH"],unrate[12:24]["VALUE"],c="blue" ,label = "1949")

plt.legend(loc = "best")#线条标签的位置，选择最好的位置，大多数会选择在空白的地方
plt.xlabel("Month Integer")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends ,1948-1949")

# print(help(plt.legend))

plt.show()
```
