```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#matplotlib 是一个可视化工具库，用于制作图表等
#pyplot是折线图类
```

```python
unrate = pd.read_csv("csv/unrate.csv")
#利用pandas to_datetime函数将日期字符串转为日期对象，替换DATE列
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
print(unrate.head(12))
```

```python
#看一下pyplot的图表
plt.plot()
plt.show()
```

```python
#取前12个切片数据，以DATE为X轴，VALUE为Y轴绘制折线图
first_twelve = unrate[0:12]
plt.plot(first_twelve["DATE"],first_twelve["VALUE"])
plt.show()
```

```python
#x坐标轴太乱，通过xticks方法指定x坐标数值绘制时旋转45度
plt.plot(first_twelve["DATE"],first_twelve["VALUE"])
plt.xticks(rotation = 45)
plt.show()
```

```python
#添加标题与单位
plt.plot(first_twelve["DATE"],first_twelve["VALUE"])
plt.xticks(rotation = 90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends ,1948")
plt.show()
```
